```python
%autoreload 2
from src import gauss_jacobi
from src import gauss_seidel

# Solución que diverge
A = [[2, 10], [3, 2]] 
b = [16, 11]
x0 = [1, 1]

solucion1, trayectoria1 = gauss_jacobi(A=A, b=b, x0=x0, tol=1e-4, max_iter=50)
solucion2, trayectoria2 = gauss_seidel(A=A, b=b, x0=x0, tol=1e-4, max_iter=50)

# Mostrar los resultados
print("Solución con Gauss-Jacobi:", solucion1)
print("Solución con Gauss-Seidel:", solucion2)
```

```python
# Graficar la trayectoria
import numpy as np
import matplotlib.pyplot as plt

array1 = np.array(trayectoria1).squeeze()
array2 = np.array(trayectoria2).squeeze()
plt.figure(figsize=(6, 4))
plt.plot(array1[:, 0], array1[:, 1], marker="o", linestyle="-", label="Trayectoria Gauss-Jacobi")
plt.scatter(solucion1[0], solucion1[1], color="red", label="Solución Gauss-Jacobi", zorder=5)

plt.plot(array2[:, 0], array2[:, 1], marker="o", linestyle="-", label="Trayectoria Gauss-Seidel")
plt.scatter(solucion2[0], solucion2[1], color="green", label="Solución Gauss-Seidel", zorder=5)
plt.title("Convergencia de Métodos Iterativos")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.grid()
plt.show()
```
