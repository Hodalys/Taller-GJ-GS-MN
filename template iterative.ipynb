```python
%autoreload 2
from src import gauss_jacobi
from src import gauss_seidel
import numpy as np

A = [[3, 2], [2, 10]] # fila 2, fila 1
b = [11, 16]
x0 = [1, 1]

# Definir la tolerancia y el número máximo de iteraciones
tol = 1e-4
max_iter = 50
solucion1, trayectoria1 = gauss_jacobi(A=A, b=b, x0=x0, tol=tol, max_iter=max_iter)
solucion2, trayectoria2 = gauss_seidel(A=A, b=b, x0=x0, tol=tol, max_iter=max_iter)

# Mostrar los resultados
print("Solución encontrada:", solucion1)
print("Solución encontrada:", solucion2)
```

```python
# Graficar la trayectoria
import matplotlib.pyplot as plt

trayectoria_array1 = np.array(trayectoria1).squeeze()
trayectoria_array2 = np.array(trayectoria2).squeeze()
plt.figure(figsize=(6, 4))
plt.plot(trayectoria_array1[:, 0], trayectoria_array1[:, 1], marker="o", linestyle="-", label="Trayectoria Gauss-Jacobi")
plt.scatter(solucion1[0], solucion1[1], color="red", label="Solución Gauss-Jacobi", zorder=5)

plt.plot(trayectoria_array2[:, 0], trayectoria_array2[:, 1], marker="o", linestyle="-", label="Trayectoria Gauss-Seidel")
plt.scatter(solucion2[0], solucion2[1], color="green", label="Solución Gauss-Seidel", zorder=5)
plt.title("Convergencia de Métodos Iterativos")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.grid()
plt.show()
```
